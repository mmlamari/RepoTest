```python
%matplotlib inline

import pandas.io.data as web
import datetime as dt
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
import scipy.stats  as st
import matplotlib.mlab as mlab
import math
```

```python
start = dt.datetime(1998, 1, 1)
end = dt.datetime(2009, 12, 31)
vbltx = web.DataReader("vbltx", 'yahoo', start, end)
fmagx = web.DataReader("fmagx", 'yahoo', start, end)
SBUX  = web.DataReader("SBUX", 'yahoo', start, end)
#SBUX = pd.DataFrame(pd.Series(SBUX), columns=['Date'])
sbux_pr=pd.DataFrame(SBUX['Adj Close'])
fmagx_pr=pd.DataFrame(fmagx['Adj Close'])
vbltx_pr=pd.DataFrame(vbltx['Adj Close'])
All_prices = pd.merge(vbltx_pr, fmagx_pr, left_index =True ,right_index=True, how = 'outer')
All_prices = pd.merge(All_prices, sbux_pr, left_index =True ,right_index=True, how = 'outer')
columns = ['vbltx_pr' , 'fmagx_pr' , 'sbux_pr']
All_prices.columns = columns
All_prices['re_vbltx'] = 0
All_prices['re_fmagx'] = 0
All_prices['re_sbux'] = 0

for i in range(1,len(All_prices)): 
    All_prices['re_vbltx'].iloc[i] = np.log(All_prices['vbltx_pr'][i]) - np.log(All_prices['vbltx_pr'][i-1]) 
    All_prices['re_sbux'].iloc[i] = np.log(All_prices['sbux_pr'][i]) - np.log(All_prices['sbux_pr'][i-1]) 
    All_prices['re_fmagx'].iloc[i] = np.log(All_prices['fmagx_pr'][i]) - np.log(All_prices['fmagx_pr'][i-1]) 

#All_prices['re_vbltx'].plot()   

plt.figure(figsize=(20,10))
ts1=np.cumsum(All_prices['re_vbltx'] )
ts2=np.cumsum(All_prices['re_fmagx'])

ts3=np.cumsum(All_prices['re_sbux'])
ts1.plot(label = 'vbltx')
ts2.plot(label='fmagx')
ts3.plot(label='sbux')
plt.legend(loc=0)
plt.show()      





#SBUX = SBUX.set_index([SBUX.index.year ,SBUX.index.month])
```
