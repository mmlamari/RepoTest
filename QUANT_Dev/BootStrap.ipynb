```python
B=999
mu=[]
nobs=len(All_prices)

for i in range(1,B):
    data  = np.random.choice(All_prices[col], size=nobs, replace=True, p=None)
    mu.append(data.mean())

print 'mean: Actual ' ,All_prices[col].mean()     
print 'mean: Boostrap ' ,np.mean(mu) 
print 'bias: ' ,np.mean(mu) - All_prices[col].mean()

print 'STD ERROR: Actual ' ,All_prices[col].std()/np.sqrt(nobs)     
print 'STD ERROR: Boostrap ' ,np.std(mu)  
print 'bias ERROR: ' , np.std(mu) - All_prices[col].std()/np.sqrt(nobs)     
    
```
